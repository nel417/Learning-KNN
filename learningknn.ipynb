```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.metrics import accuracy_score

from sklearn.model_selection import train_test_split

from scipy.stats import mode
```

```python
def euclidian_distance(pt1, pt2):
    distance = np.sqrt(np.sum(np.square(pt1-pt2)))
    return distance
```

```python
a = np.array([3,4])
b = np.array([5,9])

print(euclidian_distance(a,b))
```

```python
def KNN(X_train, X_test, Y_train, Y_test, k_val):
    y_hat = []
    for test_point in X_test.to_numpy():
        distances = []
        for i in range(len(X_train)):
            distances.append(euclidian_distance((np.array(X_train.iloc[i])), test_point))
        distance_data = pd.DataFrame(data = distances, columns=['distance'], index=Y_train.index)
        k_neighbors_list = distance_data.sort_values(by=['distance'], axis=0)[:k_val]
        labels = Y_train.loc[k_neighbors_list.index]
        voting = mode(labels).mode[0]
        y_hat.append(voting)
    return y_hat
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=42)

y_hat_test = KNN(X_train, X_test, Y_train, Y_test, k_val=5)
```

```python
print(accuracy_score(Y_test, y_hat_test))
```

```python

accuracy_vals = []
for i in range(1,15):
    y_hat_test = y_hat_test = KNN(X_train, X_test, Y_train, Y_test, k_val=i)
    accuracy_vals.append(accuracy_score(Y_test, y_hat_test))

plt.plot(range(1,15), accuracy_vals, color='blue', marker='x', linestyle='dashed')
```

```python
y_hat_test = KNN(X_train, X_test, Y_train, Y_test, k_val=5)

print(accuracy_score(Y_test, y_hat_test))

for i in range(len(y_hat_test)):
    if(y_hat_test[i] == 0):
        plt.scatter(X_test.iloc[i]['Age'], X_test.iloc[i]['Salary'], color='green')
    if(y_hat_test[i] == 1):
        plt.scatter(X_test.iloc[i]['Age'], X_test.iloc[i]['Salary'], color='red')

plt.style.use('ggplot')
plt.xlabel("age")
plt.ylabel('salary')
plt.title("KNN Results")

        
```
